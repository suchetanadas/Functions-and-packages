<div class="alert alert-block alert-warning">
    <b>Exercise 1:</b> Write a function that computes the level of output generated by a constant returns to scale Cobb-Douglas production function, i.e., such that it computes $A\cdot K^\alpha \cdot L^{1-\alpha}$.
</div>

```python
def CobbDouglas(K=1, L=1, A=1, alpha=0.3):
    '''
    This function computes the level of output of a Cobb-Douglas production function.
    Given values of K, L, A, and alpha it returns $A\cdot K^\alpha \cdot L^{1-\alpha}$.
    Parameters:
    -----------
    K: float, integer or any other number, default=1
    L: float, integer or any other number, default=1
    A: float, integer or any other number, default=1
    alpha: float between (0,1), default=0.3
    
    Returns:
    --------
    out = A * K**alpha * L**(1-alpha)
    
    Example:
    >>> out = CobbDouglas(4, 9, 2, 1/2)
    >>> out
        12
    '''
    return A * K**alpha * L**(1-alpha)
```

```python
 CobbDouglas(4, 9, 2, 1/2)
```

<div class="alert alert-block alert-warning">
    <b>Exercise 2:</b> Use the function you created and plot the production function as a function of $K$ for given values of $A$ and $L$. <b>Hint:</b> Use the <code>np.linspace(0, 5, 100)</code> function to create an array of values of $K$ with 100 point between 0 and 5.
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

K_values = np.linspace(0, 5, 100)
# Set values for A and L
A_value = 2
L_value = 9

output_values = CobbDouglas(K_values, L_value, A_value)
# Plot the production function
plt.figure(figsize=(8, 6))
plt.plot(K_values, output_values, label=f"Cobb-Douglas (A={A_value}, L={L_value})")
plt.xlabel("Capital (K)")
plt.ylabel("Output")
plt.title("Cobb-Douglas Production Function")
plt.grid(True)
plt.legend()
plt.show()
```

<div class="alert alert-block alert-warning">
    <b>Exercise 3:</b> Show with a plot the effect of increasing $A$ from 1 to 2, 4, or 5.
</div>

```python
# Create an array of K values
K_values = np.linspace(0, 5, 100)

# Set constant values for L and alpha
L_value = 9
alpha_value = 0.3

# Compute output for different A values
A_values = [1, 2, 4, 5]
output_values = [CobbDouglas(K_values, L_value, A_val, alpha_value) for A_val in A_values]

plt.figure(figsize=(8, 6))
for i, A_val in enumerate(A_values):
    plt.plot(K_values, output_values[i], label=f"Cobb-Douglas (A={A_val}, L={L_value})")

plt.xlabel("Capital (K)")
plt.ylabel("Output")
plt.title("Cobb-Douglas Production Function")
plt.grid(True)
plt.legend()
plt.show()
```

<div class="alert alert-block alert-warning">
    <b>Exercise 4:</b> Show with a plot the effect of increasing $L$ from 1 to 2, 4, 5.
</div>

```python
# Create an array of K values
K_values = np.linspace(0, 5, 100)

# Set constant values for L and alpha
A_value = 2
alpha_value = 0.3

# Compute output for different A values
L_values = [1, 2, 4, 5]
output_values = [CobbDouglas(K_values, L_value, A_val, alpha_value) for L_val in L_values]

plt.figure(figsize=(8, 6))
for i, L_val in enumerate(L_values):
    plt.plot(K_values, output_values[i], label=f"Cobb-Douglas (={L_val}, A={A_value})")

plt.xlabel("Capital (K)")
plt.ylabel("Output")
plt.title("Cobb-Douglas Production Function")
plt.grid(True)
plt.legend()
plt.show()
```
